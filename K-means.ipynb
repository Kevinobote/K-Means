# Learning K-Means

```python
import random
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.datasets._samples_generator import make_blobs
```

```python
X, y = make_blobs(n_samples=5000, centers=[[4,4],[-2, -1],[2, -3], [1, 1]], cluster_std=0.9)
```

```python
# Display the scatter plot of the generated data

plt.scatter(X[:, 0], X[:, 1], marker='*')
plt.title("Generated Data")
plt.show()
```

```python
# Initialize K-means with k_means paarameter as the output
k_means = KMeans(init = "k-means++", n_clusters = 4, n_init= 'auto')
```

```python
# import sklearn
# print(sklearn.__version__)
```

```python
# Fit the K-means model with the feature matrix we generated
k_means.fit(X)
```

```python
# Get the cluster labels

labels = k_means.labels_
labels

```

```python
# Get the coordinates of the  clusters using the k-means .cluster_centers method

centroids = k_means.cluster_centers_
centroids
```

```python
# create the visual plot

plt.scatter(X[:, 0], X[:, 1], c=labels, s=25, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], c='black', s=25, alpha=0.6)
plt.title("K-means Clustering")
plt.show()
```

```python
# Initialize the plot with the specified dimensions
fig = plt.figure(figsize=(6, 4))

# Colors use a color map, which will produce an array of colors based on labels
colors = plt.cm.Spectral(np.linspace(0, 1, len(set(labels))))

# Create a plot
ax = fig.add_subplot(1, 1, 1)

# Create a plot for each cluster
for k, col in zip(range(len(set(labels))), colors):
    # Create a list of all data points where the cluster label is k
    my_members = (labels == k)

    # Define the centroids or cluster center
    cluster_center = centroids[k]
    
    # Plot the points with color col
    ax.plot(X[my_members, 0], X[my_members, 1], 'w', markerfacecolor=col, marker='.')
    
    # Plot the centroids with a black cross
    ax.plot(cluster_center[0], cluster_center[1], '*', markerfacecolor=col, markeredgecolor='k', markersize=6)
    
# Title of the plot
ax.set_title('KMeans')

# Remove x-axis ticks
ax.set_xticks(())

# Remove y-axis ticks
ax.set_yticks(())

plt.show()
```
